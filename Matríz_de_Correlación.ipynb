```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import yfinance as yf

# 1. Extraccin de datos
symbol = 'GOOGL'
df = yf.download(symbol, start='2004-08-19', end='2024-09-20', interval='1d')# oja que carga hasta el 19 de 09 de 2024
df.reset_index(inplace=True)
df['Date'] = pd.to_datetime(df['Date'])

# Eliminar la columna de fecha para la correlacin, ya que no es una caracterstica numrica relevante
data_numeric = df.drop(columns=['Date'])

# Calcular la matriz de correlacin
correlation_matrix = data_numeric.corr()

# Visualizar la matriz de correlacin
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Matriz de Correlacin de las Caractersticas')
plt.show()
```

```python
import yfinance as yf
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Obtener los datos de la accin de Google desde Yahoo Finance
ticker = "GOOGL"  # Ticker de Google
start_date = "2004-08-19"
end_date = "2024-11-10"

# Descargar los datos
data = yf.download(ticker, start=start_date, end=end_date)

# Calcular la matriz de correlacin
correlation_matrix = data.corr()

# Visualizar la matriz de correlacin como un mapa de calor
plt.figure(figsize=(10,8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Matriz de  Correlacin de Google')
plt.show()
```

```python
# Correlacin de Spearman
spearman_corr = data_numeric.corr(method='spearman')
plt.figure(figsize=(10, 8))
sns.heatmap(spearman_corr, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Matriz de Correlacin de Spearman')
plt.show()

# Correlacin de Kendall
kendall_corr = data_numeric.corr(method='kendall')
plt.figure(figsize=(10, 8))
sns.heatmap(kendall_corr, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Matriz de Correlacin de Kendall')
plt.show()
```

```python
import yfinance as yf
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Obtener los datos de la accin de Google desde Yahoo Finance
ticker = "GOOGL"  # Ticker de Google
start_date = "2004-08-19"
end_date = "2024-11-10"

# Descargar los datos
data = yf.download(ticker, start=start_date, end=end_date)

# Calcular la matriz de correlacin
correlation_matrix = data.corr()

# Visualizar la matriz de correlacin como un mapa de calor con gradiente
plt.figure(figsize=(10,8))
# Removed the 'axis' argument as it is not valid for sns.heatmap()
# The cmap 'RdYlGn' is used for the color gradient
sns.heatmap(correlation_matrix, annot=True, cmap='RdYlGn', vmin=-1.0, vmax=1.0)
plt.title('Matriz de Correlacin de Google')
plt.show()
```
