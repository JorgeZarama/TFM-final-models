```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import yfinance as yf

def descargar_datos_accion(symbol, interval='1d'):
    """
    Descarga datos histÃ³ricos de una acciÃ³n desde Yahoo Finance, del inicio al dÃ­a de hoy.
    Analiza y visualiza outliers.
    """
    try:
        df = yf.download(symbol, period="max", interval=interval)  # Descarga datos histÃ³ricos completos
        if df.empty:
            raise ValueError("No se encontraron datos para el sÃ­mbolo especificado.")
        df.reset_index(inplace=True)
        df['Date'] = pd.to_datetime(df['Date'])

        return df
    except Exception as e:
        print(f"Error al descargar datos para {symbol}: {e}")
        return None

def analizar_outliers(df, column):
    """
    Analiza y visualiza outliers en una columna especÃ­fica usando box plots e IQR.
    """
    plt.figure(figsize=(10, 6))
    sns.boxplot(x=df[column])
    plt.title(f'Box Plot de {column} (con outliers)')
    plt.show()

    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    print(f"NÃºmero de outliers en {column}: {len(outliers)}")
    if not outliers.empty:
        print(f"Outliers en {column}:\n{outliers}")


    # VisualizaciÃ³n sin outliers (para comparaciÃ³n) â opcional
    df_no_outliers = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
    plt.figure(figsize=(10, 6))
    sns.boxplot(x=df_no_outliers[column])
    plt.title(f'Box Plot de {column} (sin outliers)')
    plt.show()




symbols = ['TSM', 'ASML', 'GOOGL', 'LRCX']

for symbol in symbols:
    print(f"\nAnalizando outliers para {symbol}:")
    df = descargar_datos_accion(symbol)

    if df is not None:
        for column in ['Open', 'High', 'Low', 'Close', 'Adj Close', 'Volume']:
            analizar_outliers(df, column)  # <---  AquÃ­ se eliminÃ³ transform_log=True
    else:
        print(f"No se pudieron descargar datos para {symbol}. Saltando...")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import yfinance as yf
!pip install statsmodels
from statsmodels import robust

def descargar_datos_accion(symbol, interval='1d'):
    """Descarga datos y maneja errores."""
    try:
        df = yf.download(symbol, period="max", interval=interval)
        if df.empty:
            raise ValueError("No se encontraron datos.")
        df.reset_index(inplace=True)
        df['Date'] = pd.to_datetime(df['Date'])
        return df
    except Exception as e:
        print(f"Error: {e}")
        return None

def analizar_outliers(df, column, threshold=3, transform_log=True):
    """Analiza outliers usando Z-score robusto con opciÃ³n de transformaciÃ³n logarÃ­tmica."""
    if transform_log:  # Aplicar transformaciÃ³n logarÃ­tmica (opcional)
        values = np.log1p(df[column])
    else:
        values = df[column]

    median = np.median(values)
    mad = robust.mad(values)
    z_scores = np.abs((values - median) / mad)

    outliers = df[z_scores > threshold]
    print(f"NÃºmero de outliers en {column}: {len(outliers)}")

    plt.figure(figsize=(10, 6))
    sns.histplot(z_scores, kde=True)
    plt.title(f'DistribuciÃ³n de Z-scores robustos para {column} (log={transform_log})')
    plt.axvline(threshold, color='red', linestyle='--', label=f'Umbral ({threshold})')
    plt.legend()
    plt.show()

    return outliers


symbols = ['TSM', 'ASML', 'GOOGL', 'LRCX']

for symbol in symbols:
    print(f"\nAnalizando outliers para {symbol}:")
    df = descargar_datos_accion(symbol)

    if df is not None:
        for column in ['Open', 'High', 'Low', 'Close', 'Adj Close', 'Volume']:
            # Analizar con transformaciÃ³n logarÃ­tmica para precios
            if column != 'Volume':  # No aplicar log a Volumen
                outliers_log = analizar_outliers(df, column, transform_log=True)
                if not outliers_log.empty:
                    print(f"Outliers (con log) en {column}:\n{outliers_log[column]}")
            else:
                # Analizar sin transformaciÃ³n logarÃ­tmica para Volumen
                outliers_no_log = analizar_outliers(df, column, transform_log=False)
                if not outliers_no_log.empty:
                    print(f"Outliers (sin log) en {column}:\n{outliers_no_log[column]}")
    else:
        print(f"No se pudieron descargar datos para {symbol}. Saltando...")
```
